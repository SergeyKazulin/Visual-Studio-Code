```python
df = usd_rate.set_index('data').join(brent_oil.set_index('date'))
```

```python
df.drop(['nominal', 'cdx'], inplace=True, axis=1)
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df['year'] = df['data'].dt.year
df['month'] = df['data'].dt.month
df['weekday'] = df['data'].dt.dayofweek
```

```python
past_days = 7
```

```python
for day in range(past_days):
    n = day + 1
    df[f'day_lag_{n}'] = df['curs'].shift(n)
    df[f'oil_lag_{n}'] = df['oil_price'].shift(n)
    df[f'mult_{n}'] = df[f'day_lag_{n}'] * df[f'oil_lag_{n}']
```

```python
df['usd_mean_week'] = df['curs'].shift(1).rolling(window=7).median()
df['oil_mean_week'] = df['oil_price'].shift(1).rolling(window=7).median()
```

```python
final_df = pd.get_dummies(df, columns=['year', 'month', 'weekday']).drop(['data', 'oil_price'], axis=1)[7:]
```

```python
X = final_df.drop('curs', axis=1)
y = final_df.curs
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
labels = ['Regressors']
LinReg = [LinReg_mae]
Ridge = [rr_mae]
Lasso = [lr_mae]
RFR = [rf_mae]
MLP = [mlp_mae]
DTR = [dtr_mae]
Elastic = [elastic_mae]
ADA = [ada_mae]


x = np.arange(len(labels))  # the label locations
width = 0.05  # the width of the bars

fig, ax = plt.subplots()

rects1 = ax.bar(x + width/1.75, LinReg, width, label='LinReg')
rects2 = ax.bar(x - width/1.75, Ridge, width, label='Ridge')
rects3 = ax.bar(x + width/- 0.58, Lasso, width, label='Lasso')
rects4 = ax.bar(x + width/ 0.58, RFR, width, label='RFR')
rects5 = ax.bar(x + width/ 0.35, MLP, width, label='MLP')
rects6 = ax.bar(x + width/ - 0.35, DTR, width, label='DTR')
rects7 = ax.bar(x + width/ 0.25, Elastic, width, label='Elastic')
rects8 = ax.bar(x + width/ -0.25, ADA, width, label='ADA')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Mean Absolute Error')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)
autolabel(rects8)


fig.tight_layout()

plt.show()
```
